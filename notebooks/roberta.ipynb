```python
from transformers import pipeline
import pandas as pd
```

```python
pipe = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-sentiment-latest")
```

```python
df = pd.read_csv('../data/raw_data.csv')
```

```python
#test the model on a sample of the data
sample_texts = df['English_Translation'].sample(20).tolist()
results = pipe(sample_texts)
for text, result in zip(sample_texts, results):
    print(f"Text: {text}\nSentiment: {result['label']}, Score: {result['score']}\n")
```
